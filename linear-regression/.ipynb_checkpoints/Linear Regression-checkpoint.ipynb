### Task 1: Fish Dataset

This one's for warmup! Load in the Fish dataset, do some basic cleaning, and run linear regression to predict fish weight!

```python
fish = pd.read_csv("fish.csv")
fish.head()
```

```python
# One-hot encode fish species
dummies = pd.get_dummies(fish.Species)
merged = pd.concat([fish, dummies], axis='columns')

# Create a new column with the mean of all 3 length measurements
merged['Length_avg'] = merged[['Length1', 'Length2', 'Length3']].mean(axis=1)
merged 

    
fish_cleaned = merged[["Bream", "Parkki", "Perch", "Pike", "Roach", "Smelt", 
                       "Whitefish", "Length_avg", "Height", "Width", "Weight"]]


# Train-test split
fish_X = fish_cleaned.drop(columns="Weight")
fish_y = fish_cleaned["Weight"]
X_train, X_test, y_train, y_test = train_test_split(fish_X, fish_y, test_size=0.2)
```

```python
# Correlation matrix
fig, ax = plt.subplots(figsize=(16,16))
dropSelf = np.zeros_like(fish_cleaned.corr())
dropSelf[np.triu_indices_from(dropSelf)] = True
sns.heatmap(fish_cleaned.corr(), annot=True, ax=ax, cmap='coolwarm', linewidths=0.5, fmt="0.2f", mask=dropSelf).set(
    title = 'Feature Correlation', xlabel = 'columns', ylabel = 'columns')

plt.show()
```

PCA code referenced from https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60

```python
# Scale data for PCA
scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Make an instance of the PCA Model, transform the data
pca = PCA(0.95)
pca.fit(X_train)

X_train = pca.transform(X_train)
X_test = pca.transform(X_test)

pca.n_components_
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
print("R^2:", reg.score(X_test, y_test))
print("Coefficients:", reg.coef_)
print("Intercept:", reg.intercept_)
```

### Task 2: FIFA 19 Dataset

This one will require a little bit more artistry to get a good score. Load in the Fifa CSV and take a look at the contents. You'll have to perform some data cleaning and feature engineering on the data (removing columns, dealing with nulls, one hot encoding, etc). Once you've completed, run linear regression and see how your model does predicting overall score! Make sure you display any tables or graphs you decided to use!

```python
fifa = pd.read_csv("fifa.csv")
pd.set_option('display.max_columns', None)
```

```python
# Remove unwanted columns and rows
fifa = fifa.drop(columns=['Unnamed: 0', 'ID', 'Name', "Photo", "Flag", "Club Logo", "Contract Valid Until",
                          "Preferred Foot", "Work Rate", "Body Type", "Loaned From",
                          "Real Face", "Jersey Number", "Joined", "GKDiving",
                          "GKHandling", "GKKicking", "GKPositioning", "GKReflexes"])

fifa = fifa[fifa["Position"] != "GK"]

pos_cols = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
            'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM','CDM',
            'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

fifa = fifa.dropna(subset=pos_cols)
```

```python
# Get avg position scores, remove individual position score columns
def avg_pos_score(row):
    
    sum = 0
    total = 0
    
    for i in pos_cols:
        if (isinstance(row[i], str)):
            sum += eval(row[i])
            total += 1

    return sum / total  

# Get avg position scores, remove individual position score columns

fifa["pos_mean"] = fifa.apply(avg_pos_score, axis=1)

fifa = fifa.drop(columns=pos_cols)
```

```python
# One-hot encode Nationality, Club, and Position columns
dummies = pd.get_dummies(fifa.Nationality)
fifa = pd.concat([fifa, dummies], axis='columns')

dummies = pd.get_dummies(fifa.Club)
fifa = pd.concat([fifa, dummies], axis='columns')

dummies = pd.get_dummies(fifa.Position)
fifa = pd.concat([fifa, dummies], axis='columns')

fifa = fifa.drop(columns=["Nationality", "Club", "Position"])
```

```python
# Convert Heights -> inches, Weights -> Pounds
def height_mean_find(list):
    inches_list = []
    nope = 0
    for a in list:
        try:
            height = a.split("'")
            feet = int(height[0])
            inches = int(height[1])
            total = ((feet * 12) + inches)
            inches_list.append(total)
        except:
            nope += 1
    return inches_list

heights_inches = height_mean_find(fifa["Height"].tolist())

height_mean = np.mean(heights_inches)

def weight_mean_find(list):
    pounds_list = []
    nope = 0
    for i in list:
        try:
            pounds = i[:-3]
            pounds_list.append(int(pounds))
        except:
            nope += 1
    return pounds_list

pounds_numbers = weight_mean_find(fifa["Weight"].tolist())
pounds_mean = np.mean(pounds_numbers)

def height_inches(row):
    try:
        a = row["Height"]
        height = a.split("'")
        feet = int(height[0])
        inches = int(height[1])
        return ((feet * 12) + inches)
    except:
        return height_mean
    
def pounds_numbers_only(row):
    try:
        i = row["Weight"]
        pounds = i[:-3]
        return int(pounds)
    except:
        return pounds_mean
    
fifa["Height Inches"] = fifa.apply(height_inches, axis = 1)
fifa["Pounds Numbers"] = fifa.apply(pounds_numbers_only, axis = 1)

fifa = fifa.drop(columns=["Height", "Weight"])
```

```python
# Clean the Wage, Value, and Release Clause columns
def to_int(df_value):
    try:
        value = float(df_value[1:-1])
        unit = df_value[-1:]
        # Millions
        if unit == 'M':
            value = value * 1000000
        # Thousands
        elif unit == 'K':
            value = value * 1000
    except:
        value = 0
    return value

fifa['Value'] = fifa['Value'].apply(to_int)
fifa['Wage'] = fifa['Wage'].apply(to_int)
fifa['Release Clause'] = fifa['Release Clause'].apply(to_int)
```

```python
len(fifa.columns)
```

```python
# Modeling
X_train, X_test, y_train, y_test = train_test_split(fifa.drop(columns="Overall"), fifa["Overall"], test_size=0.2)

scaler = StandardScaler()

# Fit on training set only.
scaler.fit(X_train)

# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# Make an instance of the PCA Model, transform the data
pca = PCA(0.95)
pca.fit(X_train)

print("PCA components:", pca.n_components_)

X_train = pca.transform(X_train)
X_test = pca.transform(X_test)

reg = LinearRegression().fit(X_train, y_train)

print("R^2:", reg.score(X_test, y_test))
print("Coefficients:", reg.coef_)
print("Intercept:", reg.intercept_)
```
